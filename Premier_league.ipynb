## Possession Analysis

```python
home_possession=df.groupby('Home Team')['home_possessions'].mean().sort_values(ascending=False)
```

```python
home_possession
```

```python
away_possession=df.groupby('Away Team')['away_possessions'].mean().sort_values(ascending=False)
```

```python
away_possession
```

```python
type(away_possession)
```

```python
possession_avg_combined=pd.merge(home_possession,away_possession,how='inner',left_on='Home Team',right_on='Away Team')
```

```python
possession_avg_combined['Team']=['Manchester City','Arsenal','Liverpool','Brighton and Hove Albion',
                                 'Chelsea','Newcastle United','Manchester United','Tottenham Hotspur',
                                 'Fulham','Aston Villa','Wolverhampton Wanderers','Leicester City', 
                                 'Leeds United','Crystal Palace','Southampton','Everton','Brentford'
                                 ,'West Ham United','Bournemouth','Nottingham Forest']
```

```python
possession_avg_combined
```

```python
plt.figure(figsize=(12, 6))
width=0.3
a=np.arange(len(possession_avg_combined['Team']))
b=list(possession_avg_combined['Team'])
plt.bar(a, possession_avg_combined['home_possessions'],width=width, label='Home', color='lightblue')
plt.bar(a+width, possession_avg_combined['away_possessions'],width=width, label='Away', color='Pink')
plt.xticks(a+.15,b,rotation=90)
plt.xlabel('Team')
plt.ylabel('Possession')
plt.title('Possession for Each Team (Home and Away)')
plt.legend()
plt.tight_layout()
plt.show()
```

## Home/Away Results

```python
df.loc[df['Goals Home']>df['Away Goals'],'result']='H'
df.loc[df['Goals Home']<df['Away Goals'],'result']='A'
df.loc[df['Goals Home']==df['Away Goals'],'result']='D'
```

### Home Analysis

```python
home_win=df[df['result']=='H'].groupby('Home Team')['result'].count()
```

```python
home_draw=df[df['result']=='D'].groupby('Home Team')['result'].count()
```

```python
home_loss=df[df['result']=='A'].groupby('Home Team')['result'].count()
```

```python
Home_Match=pd.DataFrame({'Team':home_win.index,'Home_Win':home_win.values,'Home_Draw':home_draw.values,'Home_Loss':home_loss.values})
```

```python
Home_Match
```

```python
plt.figure(figsize=(12, 6))
width=0.2
p=np.arange(len(Home_Match['Team']))
x=list(home_win.index)
plt.bar(p-width, Home_Match['Home_Win'],width=width,label='Home Win', color='lightblue')
plt.bar(p, Home_Match['Home_Draw'],width=width, label='Home Draw', color='pink')
plt.bar(p+width, Home_Match['Home_Loss'],width=width,label='Home Loss', color='y',alpha=0.5)
plt.xticks(p,x,rotation=90)
plt.xlabel('Team')
plt.ylabel('Win/Draw/Loss')
plt.title('Home Match Analysis For Each Team')
plt.legend()
plt.tight_layout()
plt.show()
```

### Away Analysis

```python
away_win=df[df['result']=='A'].groupby('Away Team')['result'].count()
```

```python
away_drow=df[df['result']=='D'].groupby('Away Team')['result'].count()
```

```python
away_loss=df[df['result']=='H'].groupby('Away Team')['result'].count()
```

```python
Away_Match=pd.DataFrame({'Team':away_win.index,'Away_Win':away_win.values,'Away_Draw':away_drow.values,'Away_Loss':away_loss.values})
```

```python
Away_Match
```

```python
plt.figure(figsize=(12, 6))
width=0.2
p1=np.arange(len(Home_Match['Team']))
x1=list(away_win.index)
plt.bar(p1-width, Away_Match['Away_Win'],width=width,label='Away Win', color='lightblue')
plt.bar(p1, Away_Match['Away_Draw'],width=width, label='Away Draw', color='pink')
plt.bar(p1+width, Away_Match['Away_Loss'],width=width,label='Away Loss', color='y',alpha=0.5)
plt.xticks(p1,x1,rotation=90)
plt.xlabel('Team')
plt.ylabel('Win/Draw/Loss')
plt.title('Away Match Analysis For Each Team')
plt.legend(loc=2)
plt.tight_layout()
plt.show()
```

## Points Calculation

```python
Match_summary=pd.merge(Home_Match,Away_Match,how='inner',on=['Team'])
```

```python
Match_summary
```

```python
Match_summary['Point']=3*(Match_summary['Home_Win']+Match_summary['Away_Win'])+Match_summary['Home_Draw']+Match_summary['Away_Draw']
```

```python
Match_summary=Match_summary.sort_values(by='Point',ascending=False).reset_index(drop=True)
```

```python
Match_summary
```

```python
plt.figure(figsize=(10, 6))
plt.bar(Match_summary['Team'],Match_summary['Point'],color=plt.cm.Paired.colors)
plt.xlabel('Teams')
plt.ylabel('Team Points')
plt.title('Points for Each Team')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
